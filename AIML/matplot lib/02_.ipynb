```python
import matplotlib.pyplot as plt                 
x=['Monday' , 'Tuesday','wed','thu','fri','sat','sun']
y=['10','20','30','60','50','20','100']
plt.plot(x,y)
plt.title('bakery sales report')
plt.xlabel('Days ')
plt.ylabel('units sold')
plt.show()
```

```python
#plt.plot(x,y color='color name',linestyle="linewidth"=value,'marker symbol','label name')   
months=[1,2,3,4]
sales=[1000,2000,3000,5000]                          
plt.plot(months,sales,color='blue',linestyle=':',linewidth=2,marker='o',label='2025 sales')
plt.xlabel('months')
plt.ylabel('units')
plt.title('Slaes report hai bahi')
plt.legend()
plt.grid(color='gray',linestyle=':',linewidth=2)
# plt.xlim(1,2)
# plt.ylim(0,3000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```

```python
#plt.bar(x,height,color='color name',width=value,label='label name')
import matplotlib.pyplot as plt                 
product=['A','B','C','D']
sales=[1000,1200,300,1500]
#plt.bar is vertical bar graph ploting
#plt.bar(product,sales,color="black",label='sales 2025')
#plt.barh(is for plotting the data horzontally)
plt.barh(product,sales,color="black",label='sales 2025')
plt.title('Sales report')
plt.ylabel('units sold')
plt.xlabel('product varity')
plt.yticks(['A','B','C','D'],['Hearts','Livers','kidney','eyes'])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt 
#plt.pie(value , labels='label list,color='given color',autopct='%1.1f%')
region=['North','south','East','West']
revenue=[5500,5000,3000,4500]
plt.pie(revenue,labels=region,autopct='%1.1f%%',colors=['gold','pink','blue','purple'])
plt.title('revenue percentage of of country India ')
plt.show()

   
```

```python
import matplotlib.pyplot as plt 
x=[1,2,3,4]
y=[10,20,15,25]

plt.subplot(1,2,1)#1row,2colomn,1st subplot
plt.plot(x,y)
plt.title('line chart')
plt.subplot(1,2,2)# 1 row,2 column,2ndsubplot
plt.bar(x,y)
plt.title('bar chart')
plt.grid()
plt.tight_layout()
plt.show()
```
