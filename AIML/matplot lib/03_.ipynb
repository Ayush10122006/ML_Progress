```python
import matplotlib.pyplot as plt 

# Create 2 subplots (1 row, 2 columns)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

# Line plot
ax[0].plot(x, y)
ax[0].set_title('Line plot')

# Bar chart
ax[1].bar(x, y)
ax[1].set_title('Bar chart')

plt.show()
```
